```python
from abc import ABC, abstractmethod
from tkinter import *
from tkinter import ttk

#################### DATA ##################
faith_meals = {
    "Fried Rice": 400,
    "White Rice and Stew": 400,
    "Jollof Rice": 400,
    "Beans": 200,
    "Chicken": 1000
}

coop_meals = {
    "Jollof Rice and Stew": 200,
    "White Rice and Stew": 200,
    "Fried Rice": 200,
    "Salad": 100,
    "Plantain": 100
}

stu_centre_meals = {
    "Fried RIce and Chicken": 800,
    "Pomo Sauce": 300,
    "Spaghetti Jollof": 500,
    "Amala/Ewedu": 500,
    "Semo with Eforiro Soup": 500
}

################## CLASSES ########################

class External_Vendors(ABC):
    @abstractmethod
    def menu(self):
        pass

class Vendor(External_Vendors):
    meals = None
    def menu(self):
        return self.meals

class Faith(Vendor):
    def __init__(self):
        super().__init__()
        self.meals = faith_meals


class Cooperative(Vendor):
    def __init__(self):
        super().__init__()
        self.meals = coop_meals


class StudentCentre(Vendor):
    def __init__(self):
        super().__init__()
        self.meals = stu_centre_meals

    
################### GUI ##########################

####### Sumbit function #########
def view_menu():
    title = None
    menu = None

    vendor = vendor_entry.get().lower()

    if vendor == "faith hostel":
        vendor = Faith()
        title = "Faith Hostel"
        menu = vendor.meals

    elif vendor == "cooperative hostel":
        title = "Cooperative Hostel"
        vendor = Cooperative()

    elif vendor == "student centre":
        title = "Student Centre"
        vendor = StudentCentre()

    menu = vendor.menu()

    window = Toplevel(root)
    window.geometry("400x300")
    window.title(f"{title} Menu")

    #Insert all the meals and their respective prices into a table.

    table = ttk.Treeview(window, columns= ("meal", "price"), show = "headings")
    table.heading("meal", text="Main Meal")
    table.heading("price", text= "Price(N)")
    table.pack(fill="both", expand=True)

    for meal in menu:
        data = (meal, menu[meal])
        table.insert(parent="", index=END, values=data)


########## Main window: root ##############

root = Tk()
root.geometry("300x300")
root.title("External Vendors Cafeteria")

Label(text="Select Cafeteria").pack()


vendors = ["Faith Hostel", "Cooperative Hostel", "Student Centre"]
vendor_entry = ttk.Combobox(root, values=vendors)
vendor_entry.pack()

sumbit_button = Button(root, text="View Menu", command=view_menu)
sumbit_button.pack()

root.mainloop()


```
