A PYTHON GUI PROGRAM THAT SHOWS THE SERVICES OF SELECTED DIVISIONS OF ZENITH BANK USING INHERITANCE AND POLYMORPHISM CONCEPTS OF OOP.

```python
from tkinter import *
from tkinter import ttk

services = [
    
    "Lines of credit",
    "Investment management and accounts",
    "Insurance",
    "Retirement and education accounts",
    "loans and mortgagges",
    "Checking and saving",

    "Multi-currency management services and products",
    "Foreign currency accounts",
    "Foreign currency credit cards",
    "Transborder advisory services",
    "Liquidity management",

    "Lines of credit",
    "Investment management and accounts",
    "Insurance",
    "Advisory services"
]

class Zenith:

    def __init__(self):
        self.services = services

    def unique_services(self):
        unique_services = []

        for service in self.services:
            if services.count(service) == 1:
                unique_services.append(service)

        return unique_services

    def mutual_services(self):
        mutual_services = []

        for service in self.services:
            if services.count(service) > 1:
                mutual_services.append(service)

        return mutual_services


class RetailBanking(Zenith):
    def __init__(self):
        self.services = services[:6]


class GlobalBanking(Zenith):
    def __init__(self):
        self.services = services[6:11]



class CommercialBanking(Zenith):
    def __init__(self):
        self.services = services[11:]



def display_services():
    division = division_entry.get()

    #Create division object
    if division == "Retail Banking":
        division = RetailBanking()

    elif division == "Global Banking":
        division = GlobalBanking()

    elif division == "CommercialBanking":
        division = CommercialBanking()

    d_unique_services = division.unique_services()
    d_mutual_services = division.mutual_services()

    #New window
    window = Toplevel(root)
    window.geometry("500x500")
    window.title("Services")


    #Display Services
    Label(window,text="Unique Services:", font=("Arial", 15, "bold", "underline")).pack()
    for service in d_unique_services:
        Label(window,text=service).pack()

    Label(window,text="Mutual Services:", font=("Arial", 15, "bold", "underline")).pack()

    if d_mutual_services != []:
        for service in d_mutual_services:
            Label(window,text=service).pack()
    else:
        Label(window, text="None").pack()



root = Tk()
root.geometry("500x500")
root.title("Zenith Bank Login")



Label(text="Surname").pack()

surname_entry = Entry()
surname_entry.pack()


Label(text= "First Name").pack()
firstname_entry = Entry()
firstname_entry.pack()


Label(text= "Division").pack()
divisions = ["Retail Banking", "Global Banking", "Commercial Banking"]
division_entry = ttk.Combobox(values=divisions)
division_entry.pack()


submit_button = Button(text="Check Division Services", command=display_services)
submit_button.pack()


root.mainloop()
```
