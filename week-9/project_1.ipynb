A PYTHON LOGIN PROGRAM AN EMPLOYEE IF HIS/HER NAME IN THE LIST OF EMPLOYEES AND ASSIGN THEM THEIR TASK OF THE DAY ELSE DENIES ACCESS
MAKING USE OF TKINTER AND OOP

```python

from random import choice
from tkinter import *
from tkinter import messagebox

class Employee:

    def __init__(self, name):
        self.name = name
        self.task = None
        self.attendence = False

    def check_employee(self):
        return  self.name in employees

    def take_attendance(self):
        self.attendence = True
        employees_in_attendance.append(self)

    def assign_task(self):
        self.task = choice(tasks)

    def refuse_access(self):
        messagebox.showerror("Access Denied", "Sorry this name is not on our employees' database")


employees = [("mary","evans"), ("eyo","ishan"), ("durojaiye" ,"dare"), ("adams","ali"), ("andrew","ugwu"), ("stella","mankinde"), 
             ("jane", "akibo"), ("ago","james"), ("michell" ,"taiwo"), ("abraham" ,"jones") , ("nicole", "anide"), ("kosi", "korso"), 
             ("adele" "martins"), ("emmanuel", "ojo"), ("ajayi","fatima")]

tasks = ["Loading", "Transporting", "Reveiwing Orders", "Customer Service", "Delivering Items"]

#list of employees in attendance
employees_in_attendance = []



from tkinter import *

def login():
    name = (surname_entry.get().lower(), firstname_entry.get().lower())

    if name[0] == "" or name[1] == "":
        messagebox.showerror("Input Error", "Invalid Input.\n Entries must not be empty.")

    else:

        employee = Employee(name)

        if employee.check_employee():

            employee.assign_task()
            employee.take_attendance()

            window = Toplevel(root)
            window.geometry("300x200")
            window.title("Welcome")

            Label(window, text=f" Welcome {name[0].upper()} {name[1].upper()}\n Your task today is {employee.task}.").pack()

        else:
            employee.refuse_access()


root = Tk()
root.geometry("300x200")
root.title("Employee Login")

Label(text="INPUT YOUR NAME").place(x=100, y=0)
Label(text="Surname:").place(x=10, y=20)
Label(text="First Name:").place(x=10, y= 40)

surname_entry = Entry()
surname_entry.place(x=100, y=20)
firstname_entry = Entry()
firstname_entry.place(x=100, y=40)


login_button = Button(text="Login", command=login)
login_button.place(x=150, y=70)



root.mainloop()
```
